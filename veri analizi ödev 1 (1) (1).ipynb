```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Mustafa\Downloads\KAMP ÖDEV\50_Startups.csv")
df.head()
```

SORU 1 : R&D harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterin

```python
plt.figure(figsize=(6,4))
plt.scatter(df["R&D Spend"], df["Profit"], color = "pink")
plt.title("The Relationship Between R&D Spend and Profit")
plt.xlabel("R&D Spend") 
plt.ylabel("Profit")
plt.show()
```

Soru 2 : Yönetim harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterin.

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Administration"], df["Profit"], color = "yellow")
plt.title("The Relationship Between Administration and Profit")
plt.xlabel("Administration") 
plt.ylabel("Profit")
plt.show()
```

Soru 3: Eyaletlere göre ortalama kârları bar chart ile karşılaştırın.

```python
df1 = df.groupby("State")["Profit"].mean().reset_index()
# groupby(verilen başlık içerisinde aynı olan ifadeleri alır)[sadece belirtilen sütunu seçer].mean().reset_index()
#mean() ortalama alır
#reset_index() ise tablo yani dataframe şeklinde sonucu döndürür.
df1.head()
```

```python
plt.figure(figsize=(6,4))
plt.bar(df1["State"], df1["Profit"])
plt.title("The Relationship Between States and Profits")
plt.xlabel("States")
plt.ylabel("Profit")
plt.show()
```

```python
#tüm veri setim df değişkeninde kayıtlı 
plt.figure(figsize=(6,4))  # Grafiğin boyutu, önce her zaman bununla başlanır.
# Boxplot çizimi
plt.boxplot(
    [df["R&D Spend"], df["Administration"], df["Marketing Spend"], df["Profit"]],  #state sütunu str olduğundan kullanılamaz
    patch_artist=True,
    boxprops=dict(facecolor="skyblue")  # facecolor bir key olarak kullanılır.
)
#verilen ilk arguman liste olur. Liste elemanları, tablo başlıkları olur
# Eksen etiketleri
plt.xticks(
    [1, 2, 3, 4],
    ["R&D Spend", "Administration", "Marketing Spend", "Profit"] #2.arguman tek bir liste olmalı
)

# Başlık
plt.title("R&D, Yönetim ve Pazarlama Harcamalarının Dağılımı")

# Göster
plt.show()
```
